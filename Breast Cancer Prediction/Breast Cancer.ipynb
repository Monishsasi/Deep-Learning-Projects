```python
# Importing dependencies

import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn.datasets as ds
```

```python
#  Loading dataset into a variable
df = ds.load_breast_cancer()
df
```

```python
#  Creating dataframe from the variable
cancer_data = pd.DataFrame(df.data,columns=df.feature_names)
cancer_data['Target'] = df.target
```

```python
cancer_data
```

```python
# Shows the dimension of the dataframe
cancer_data.shape
```

```python
#  Distribution of target variables
cancer_data['Target'].value_counts()
```

```python
#  Information about the dataframe
cancer_data.info()
```

```python
#  Statistical informations about dataframe
cancer_data.describe()
```

```python
#  Grouping by gender
cancer_data.groupby('Target').mean()
```

```python
X = cancer_data.drop('Target',axis=1)
y = cancer_data['Target']
```

```python
X
```

```python
y
```

```python
#  Splitting into training data and testing data
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=2)
```

```python
#  Printing shapes of each array
print(X_train.shape,X_test.shape,y_train.shape,y_test.shape)
```

```python
#  Standardise the data

from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
X_train_std = scaler.fit_transform(X_train)
X_test_std = scaler.transform(X_test)
```

```python
print(X_train_std)
print(X_test_std)
```

**Building the neural network**

```python
# Importing dependencies
import tensorflow as tf
tf.random.set_seed(3)
from tensorflow import keras
```

```python
#  Setting up layers of neural network

model = keras.Sequential([keras.layers.Flatten(input_shape=(30,)),  # converts inputs into single dimensional array 
                          keras.layers.Dense(20,activation='relu'),
                          keras.layers.Dense(2,activation='sigmoid')
                         ])
```

```python
#  Compiling the neural network
model.compile(
    optimizer="adam",
    loss="sparse_categorical_crossentropy",
    metrics = ['accuracy']
)
```

```python
#  Training the neural network
history = model.fit(X_train_std,y_train,validation_split=0.1,epochs=20)
```

```python
history.history.keys()
```

```python
#  Visualising accuracy and loss
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])

plt.title("Model Accuracy")
plt.xlabel("epoch")
plt.ylabel("accuracy")


plt.legend(['training data','validation data'],loc='lower right')
plt.show()
```

```python
#  Visualising accuracy and loss
plt.plot(history.history['loss'])
plt.plot(history.history['val_accuracy'])

plt.title("Model Loss")
plt.xlabel("epoch")
plt.ylabel("loss")


plt.legend(['training data','validation data'],loc='right')
plt.show()
```

```python
#  Accuracy of the model in the test data

loss,accuracy = model.evaluate(X_test_std,y_test)
accuracy
```
